```python
#replace space in column name
cars.columns = cars.columns.str.replace(' ','')
```

```python
# checking min & max value of a feature of dataset using function

def min_max_values(col):
    #the function takes the column name as the argument
    #and returns the top & bottom observation in that dataframe
    
    top = cars[col].idxmax()
    top_obs = pd.DataFrame(cars.loc[top])
    
    bottom = cars[col].idxmin()
    bot_obs = pd.DataFrame(cars.loc[bottom])
    
    min_max_obs = pd.concat([top_obs, bot_obs], axis=1)
    
    return min_max_obs
```

```python
#Lets check
min_max_values('Horsepower')
```

```python
#Correlation with the variable if interest
cars_corr = cars_num.corr()['Fuel_efficiency'][:-1]
cars_corr
```

```python
#Correlation Plots using 'pairplot'
for i in range(0, len(cars_num.columns),4):
    sns.pairplot(cars_num, y_vars=['Fuel_efficiency'], x_vars = cars_num.columns[i:i+4])
```

```python
#Plotting significant correlation in one plot (heatmap)

corr = cars_num.drop('Fuel_efficiency', axis=1).corr()
sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
           cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
           annot=True, annot_kws={"size": 8}, square=True);
```
